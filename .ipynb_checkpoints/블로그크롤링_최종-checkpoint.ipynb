```python
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup as bs    
import requests as req
import pandas as pd
from tqdm import tqdm_notebook
import time
```

```python
driver = wb.Chrome()
query = "부산여행"
url = "https://search.naver.com/search.naver?where=post&query={}".format(query) #블로그 주소
driver.get(url)
time.sleep(2)
title_list = []
content_list = []
for i in tqdm_notebook(range(1,500)):
    article_raw = driver.find_element_by_css_selector('#sp_blog_{} > div > a.api_txt_lines.total_tit'.format(i)) #블로그 제목
    title_list.append(article_raw.text)
    content = driver.find_element_by_css_selector('#sp_blog_{} > div > div.total_group > div > a > div'.format(i))
    content_list.append(content.text)   
    time.sleep(1)
    if i%20==0:
        driver.find_element_by_tag_name('body').send_keys(Keys.END)
        time.sleep(1)
    
```

```python
print(len(title_list))  
```

```python
try : 
    title.to_csv("블로그제목.csv",encoding='cp949')
except:
    print('이모티콘글 제거')
```

# 블로그 워드클라우드

```python
import matplotlib.pyplot as plt

from matplotlib import font_manager, rc
import sys
from wordcloud import WordCloud
import platform
```

```python
blog_word = pd.read_csv('블로그.csv',encoding="CP949")
blog_word = blog_word.iloc[:,-1]
blog_word
```

```python
stop_words = ['맛집','이','에','을','가','도','와','1','저','다','를','들','은','과','그','홍','타고','편','박','리','마을','죽','유엔','곳','성당','송도','이번','중','것','날','수']
blog_word_tag = []
blog_word_tag = [word for word in blog_word if word not in stop_words]

from collections import Counter 
count = Counter(blog_word_tag) 
blog_common_tag_30 = count.most_common(30)
```

```python
if platform.system() == 'Windows':   
    font_path = "c:/Windows/Fonts/malgun.ttf"
elif platform.system() == "Darwin":  
    font_path = "/Users/$USER/Library/Fonts/AppleGothic.ttf"
```

```python
wordcloud=WordCloud(font_path= font_path, 
                    background_color="white",
                    max_words=30,
                    relative_scaling= 0.3,
                    width = 800,
                    height = 400
                 ).generate_from_frequencies(dict(blog_common_tag_30))  
plt.figure(figsize=(15,10))
plt.imshow(wordcloud)
plt.axis('off')
# plt.savefig('C:/Users/Desktop/gbtriptag-wordcloud.png')  
```
