# 인스타그램 크롤링

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import re
from tqdm import tqdm_notebook
# 크롬 브라우저 열기
driver = webdriver.Chrome() 
url = "https://www.instagram.com/"


driver.get(url)
time.sleep(2)
email = '아이디' 
input_id = driver.find_elements_by_css_selector('input._2hvTZ.pexuQ.zyHYP')[0]
input_id.clear()
input_id.send_keys(email)

password = '비밀번호' 
input_pw = driver.find_elements_by_css_selector('input._2hvTZ.pexuQ.zyHYP')[1]
input_pw.clear()
input_pw.send_keys(password)
input_pw.submit()
time.sleep(5)


button1 = driver.find_element_by_css_selector("#react-root > section > main > div > div > div > div > button")
button1.click()



button = driver.find_element_by_css_selector("body > div.RnEpo.Yx5HN > div > div > div > div.mt3GC > button.aOOlW.HoLwm")
button.click()

def insta_searching(word):
    url = "https://www.instagram.com/explore/tags/" + word
    return url
word = "부산여행"
url = insta_searching(word)
driver.get(url)
time.sleep(3)


def select_first(driver):
    #첫번째 거 클릭
    first = driver.find_element_by_css_selector("a > div.eLAPa > div._9AhH0")
    first.click()
    time.sleep(3)
    
select_first(driver)
instagram_tags = [] 
def get_content(driver):
    #내용 얻는 함수
    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')
    # 본문 내용 
    try:
        content = soup.select('div.C4VMK > span')[0].text
    except:
        content = ' '
    # 해시태그 
    tags = re.findall(r'#[^\s#,\\]+', content)  
    tag = ''.join(tags).replace("#"," ").split() # "#" 제거 
    
    for tag_one in tag: 
        instagram_tags.append(tag_one)

    # 작성일자 
    date = soup.select('time._1o9PC.Nzb55')[0]['datetime'][:10]
    # 좋아요 
    try:
        like = soup.select('div.eo2As > section.EDfFK.ygqzn > div > div > a')[0].text[4:-1]   
    except:
        like = 0
    # 위치
    try: 
        place = soup.select('div.o-MQd > div.M30cS > div.JF9hh > a')[0].text
    except:
        place = ''
   
    data = [tag, date, like, place]
    return data

def move_next(driver):
    #다음페이지넘어가는 함수
    right = driver.find_element_by_css_selector ('a.coreSpriteRightPaginationArrow')
    right.click()
    time.sleep(3)


results=[]

for i in tqdm_notebook(range(500)):
    try:
        data = get_content(driver)  
        results.append(data)
        move_next(driver)
    except:
        time.sleep(2)
        move_next(driver)   
    
    
#해시태그 날짜 좋아요 주소
# driver.close()
```

```python
stop_words = ['부산','부산여행','여행','맞팔','평일','주말','해외여행','제주도맛집','광안리호빠','해운대호빠','부산호빠','서면호빠','부산클럽','fff','동래호빠','부산호스트바','부산호스트빠','서면클럽','ㅂㅜㅅㅏㄴ','부산맛집명성오리불고기']  #중복되는 거나 필요없는 단어 정리

instagram_tags = [word for word in instagram_tags if word not in stop_words]

from collections import Counter 
count = Counter(instagram_tags) 
common_tag_50 = count.most_common(50)

```

```python
common_tag_50
```

```python
if platform.system() == 'Windows':   
    font_path = "c:/Windows/Fonts/malgun.ttf"
elif platform.system() == "Darwin":  
    font_path = "/Users/$USER/Library/Fonts/AppleGothic.ttf"
```

```python
wordcloud=WordCloud(font_path= font_path, 
                    background_color="white",
                    max_words=30,
                    relative_scaling= 0.3,
                    width = 800,
                    height = 400
                 ).generate_from_frequencies(dict(common_tag_50))  
plt.figure(figsize=(15,10))
plt.imshow(wordcloud)
plt.axis('off')
# plt.savefig('C:/Users/Desktop/gbtriptag-wordcloud.png')  
```
