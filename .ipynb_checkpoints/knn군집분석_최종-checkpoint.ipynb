```python
import numpy as np  
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pickle
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

```

```python
trip = pd.read_csv('data.csv',encoding='cp949')
trip.head()
```

```python
trip['성별'] = trip['성별'].map({'남자':1, '여자':0})
trip['여행에서얼마나이용'] = trip['여행에서얼마나이용'].map({'1시간 이상 2시간 이하':0,'2시간 이상':1, '30분 이상 1시간 이하':2,'30분 이하':3})
trip['누구와여행'] = trip['누구와여행'].map({'가족':0,'부부/연인':1,'친구/모임':2,'홀로':3})
trip['여행우선순위'] = trip['여행우선순위'].map({ '맛_여행':0,'액티비티/레저여행':1,'체험/관광/역사':2,'축제/공연/문화예술여행':3,'휴식/힐링여행':4})
```

```python
trip.head(50)
```

```python
trip = trip.fillna(0)
trip = trip.astype(int)
```

```python
km = KMeans(n_clusters=5, 
            init='k-means++', 
            n_init=10, 
            max_iter=300,
            random_state=1)
y_km = km.fit_predict(trip)
```

```python
y_km
```

```python
cluster = pd.DataFrame(y_km)
cluster.columns = ['cluster']
cluster
```

```python
# cluster.to_csv('cluster_number.csv',encoding='cp949')
```

```python
from matplotlib import cm
from sklearn.metrics import silhouette_samples

cluster_labels = np.unique(y_km)
n_clusters = cluster_labels.shape[0]
silhouette_vals = silhouette_samples(trip, y_km, metric='euclidean')
y_ax_lower, y_ax_upper = 0, 0
yticks = []
for i, c in enumerate(cluster_labels):
    c_silhouette_vals = silhouette_vals[y_km == c]
    c_silhouette_vals.sort()
    y_ax_upper += len(c_silhouette_vals)
    color = cm.jet(float(i) / n_clusters)
    plt.barh(range(y_ax_lower, y_ax_upper), c_silhouette_vals, height=1.0, 
             edgecolor='none', color=color)

    yticks.append((y_ax_lower + y_ax_upper) / 2.)
    y_ax_lower += len(c_silhouette_vals)
    
silhouette_avg = np.mean(silhouette_vals)
plt.axvline(silhouette_avg, color="red", linestyle="--") 

plt.yticks(yticks, cluster_labels + 1)
plt.ylabel('Cluster')
plt.xlabel('Silhouette coefficient')

plt.tight_layout()
plt.show()
```

```python
trip_cluster = pd.concat([trip,cluster],axis=1)
```

```python
trip_cluster = trip_cluster.astype(int)
```

```python
import numpy as np
from sklearn.manifold import TSNE

# 2개의 차원으로 축소
transformed = TSNE(n_components=2).fit_transform(trip_cluster)
transformed.shape
```

```python
xs = transformed[:,0]
ys = transformed[:,1]
plt.scatter(xs,ys, c=trip_cluster['cluster'])  #라벨은 색상으로 분류됨

plt.show()
```

```python
X = trip_cluster.iloc[:,:-1]
y = trip_cluster.iloc[:,-1]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, random_state=42)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,y_train)
knn.score(X_test,y_test)
```

```python
with open("knn_model.pkl",'wb') as f :
    pickle.dump(knn,f)
```
